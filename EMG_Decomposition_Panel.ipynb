```python
import scipy.io as scio
import os

def import_Mat_file(filePath: str):

    if filePath[-4:] != ".mat":

        filePath = "".join([filePath, ".mat"])

    matFile = scio.loadmat(filePath)

    f_sampling = matFile["SamplingFrequency"][0, 0]

    time_samples = matFile["Time"][0, 0].reshape(-1)

    sEMG = matFile["Data"][0, 0]

    if sEMG.shape[1] < sEMG.shape[0]:
        sEMG = sEMG.T

    return f_sampling, time_samples, sEMG

f_sampling, time_samples, sEMG = import_Mat_file("data/First_sEMG_Signals.mat")
```

```python
def menu_parameters(n_channels:int):
    style = {'description_width': 'initial'}

    w_M = ipywi.IntSlider(min = n_channels, max = 500, step = 1, value = 120, description='Number of Iterations (M): ',style=style)
    # w_M = ipywi.HBox([ipywi.Label(value='Number of Iterations (M): '), ipywi.IntSlider(min = n_channels, max = 500, step = 1, value = 120)])
    w_R = ipywi.IntSlider(min = 4, max = 32, step = 1, value = 10, description='Extension Factor (R): ',style=style)
    # w_R = ipywi.HBox([ipywi.Label(value='Extension Factor (R): '), ipywi.IntSlider(min = 4, max = 32, step = 1, value = 10)])
    w_in = ipywi.IntSlider(min = 20, max = 60, step = 1, value = 30, description='Internal Loop Iterations: ',style=style)
    # w_in = ipywi.HBox([ipywi.Label(value='Internal Loop Iterations: '), ipywi.IntSlider(min = 20, max = 60, step = 1, value = 30)])
    w_Tolx = ipywi.FloatSlider(min = 0.00001, max = 0.01, step = 0.00001, value = 0.0023, description='FastICA Convergence: ',style=style,readout_format='.5f')
    # w_Tolx = ipywi.HBox([ipywi.Label(value='FastICA Convergence: '), ipywi.FloatSlider(min = 0.00001, max = 0.01, step = 0.00001, value = 0.0023, readout_format='.5f')])
    w_Sil = ipywi.FloatSlider(min = 0, max = 100, step = 0.1, value = 90, description='SIL Threshold (SIL [%]): ',style=style, readout_format='.1f')
    # w_Sil = ipywi.HBox([ipywi.Label(value='Silhouette Threshold (SIL [%]): '), ipywi.FloatSlider(min = 0, max = 100, step = 0.1, value = 90, readout_format='.1f')])
    w_cf = ipywi.HBox([
        ipywi.Label(value='Contrast Function g(x):'),
        ipywi.RadioButtons(
            options=[
                'x²',
                'x³',
                'log(x² + 1)',
                'tanh(x)',
                '(-x) exp(-x² / 2)'
            ],
            layout={'width': 'max-content'},
            )
    ],)
    w_winit = ipywi.HBox([
        ipywi.Label(value='Initializations of wⱼ:'),
        ipywi.RadioButtons(
            options=[
                'Random',
                'Maximum',
                'Median',
                '80th Percentile',
            ],
            layout={'width': 'max-content'},
            )
    ],)

    choosen_par = {'M':w_M,'R':w_R,'Max_inte':w_in,'Tolx':w_Tolx,'Sil':w_Sil,'Con_Fun': w_cf,'w_Initi':w_winit}
    
    return ipywi.VBox([w_M,w_R,w_in,w_Tolx,w_Sil,w_cf,w_winit]), choosen_par
```

```python
ui_menu_par, decompose_paramenters = menu_parameters(64)
display(
    ipywi.HBox(
        [ui_menu_par,
        ipywi.VBox([
            ipywi.Button(
                description='Click me',
                disabled=False,
                button_style='', # 'success', 'info', 'warning', 'danger' or ''
                tooltip='Click me',
                icon='check' # (FontAwesome names without the `fa-` prefix)
            ),
        ],),
    ],),
)
```
